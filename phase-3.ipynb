```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# here we create a sample data set for practice , we also take a csv file or excel file data 

sales_data= np.array([
    [1, 150000, 180000 , 220000 ,250000],# paradise biryani
    [2, 120000, 140000 , 160000 ,190000],# beijing bites
    [3, 200000, 230000 , 260000 ,300000],# pizza hub 
    [4, 180000, 210000 , 240000 ,270000],# Burger point 
    [5, 160000, 185000 , 205000 ,230000],# chai point 
])

print("==== Zomato sales analysis =====")
print("\n Sales data shape ",sales_data.shape)
print("\n sales data size ",sales_data.size)
print("\n i take 4th row :", sales_data[3])
print("i want the data of 1st 3 resturant :",sales_data[0:3])#we also write [:3] insted of [0:3]
print("i want the data of 1st 3 resturant :",sales_data[0:3,1:3])
```

total sales per year

```python
print(np.sum(sales_data, axis=0))#this give the sum of  column 

yearly_total=np.sum(sales_data[:,1:],axis=0)#for sum of the all columns excluding 0 th column.
print(yearly_total)

print(np.sum(sales_data[0:1,],axis=1))#for sum of the  single row 
```

```python
#MINIMUM SALES OF PER RESTURANT 
min_sales=np.min(sales_data[:,1:],axis=1)
print("minimum sales of per resturant :",min_sales)

max_sales=np.max(sales_data[:,1:],axis=1)
print("maximum sales of per resturant :",max_sales)


sum_min_sales=np.sum(np.min(sales_data[:,1:],axis=1))# heavy logic used here by self 
print("sum of all the minimum sales of per resturant :",sum_min_sales)

## average sales per restaurant 
avg_sales = np.mean(sales_data[:,1:], axis=1)
print("average sales of per resturant:",avg_sales)
```

 A cumulative sum, also known as a running total, is the sum of a sequence of numbers that is updated each time a new number is added to the sequence. It involves adding the value of the new number to the previous cumulative sum. In simpler terms, it's the total sum of a series of numbers up to a specific point. 

```python
cumsum= np.cumsum(sales_data[:,1:], axis=1)
print(cumsum)

#plot a graph of average of the cumsum data.

plt.figure(figsize=(9,6))
plt.plot(np.mean(cumsum ,axis=0))
plt.title("average of cummulative sales accross all the resturant ")
plt.xlabel("Years")
plt.ylabel("sales")
plt.grid(True)
plt.show()
```

## vector class 

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])

print("vector addition:  ", vector1 + vector2)
print("\n vector multiplication:  ", vector1 * vector2)
print("\n Dot product", np.dot(vector1,vector2))

angle = np.arccos(np.dot(vector1,vector2)/(np.linalg.norm(vector1) * (np.linalg.norm(vector2))))
print(angle)
                  
```
